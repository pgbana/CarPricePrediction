# 3. Exploring data
--------------------------------

```python
%matplotlib inline 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
cdf = pd.read_csv('cdf.csv', index_col=0)
```

----------------------
## 3.1 Relation between price and other features
---------------------

### 3.1.1 Price - Age relation and Price - Mileage relation

```python
plt.subplots(figsize=(9, 4))
plt.subplot(121)
ax = sns.kdeplot(cdf['Age'], cdf['Price']/1000, cmap="hot", shade=True, shade_lowest=True, n_levels=50)
plt.xlim(0, 25)
plt.ylim(0, 70)
plt.title('Price-Age relation')
plt.subplot(122)
ax = sns.kdeplot(cdf['Mileage']/1000, cdf['Price']/1000, cmap="hot", shade=True, shade_lowest=True, n_levels=50)
plt.xlim(0, 300)
plt.ylim(0, 70)
plt.title('Price-Mileage relation')
```

**Conclusions:** 
- There is a strict non-linear correlearion between age and price of a car. 
- The most cars are 9-15 years. 
- On the plot some subgroups are visible.
- There is not simple relation between Mileage and Price.
- On the plot, three groups are visible. 
- The common mileage for the cars lies between 125k and 250k

### 3.1.2 Relation between Mileage per year and Age - Mileage relation

```python
plt.subplots(figsize=(9, 4))
plt.subplot(121)
ax = sns.kdeplot(cdf['Mileage']/(cdf['Age']+1)/1000, cdf['Price']/1000, cmap="hot", shade=True, shade_lowest=True, n_levels=50)
plt.xlim(0, 40)
plt.ylim(0, 70)
plt.title('Mileage per year - Price relation')
plt.subplot(122)
ax = sns.kdeplot(cdf['Mileage']/1000, cdf['Age'], cmap="hot", shade=True, shade_lowest=True, n_levels=50)
plt.xlim(0, 300)
plt.ylim(0, 25)
plt.title('Mileage-Age relation')
```

**Conclusions:** 
- The typical value of Mileage per year is located close to 13k km per year.
- The mileage per year of oldest cars is the order of 10k km per year
- There is a group of cars with high price and larger mileage per year.
- There is a group of young cars with relativle large mileage.
- The company cars, propably, belongs to this two groups.
