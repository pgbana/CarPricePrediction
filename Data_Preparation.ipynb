## 2.3 Data transformation and feature selection
------------------------

New DataFrame with clean data: cdf

```python
cdf = pd.DataFrame()
```

1) Price in thousands of PLN 

```python
cdf['Price'] = cars['Cena'].astype('int32')
```

2)  Seller: private person. Some cars are sold by private person, other by companies.

```python
cdf['PrivPer'] = cars['Oferta od'].replace({'Firmy': 0, 'Osoby prywatnej':1}).astype('uint8')
```

3) Version. Ople Astra occurs in following versions: F, G, H, J, K

```python
cdf[['F','G','H','J','K']] = pd.get_dummies(cars['Wersja'].str[0])
```

4) Age. (Age = 2018 - Year of the production)

```python
cdf['Age'] = 2019 - cars['Rok produkcji'].astype('int16')
```

5) Mileage of car in km

```python
cdf['Mileage'] = cars['Przebieg'].str.replace(' ', '').str.replace('km','').astype('int32')
```

6) Engine displacement in cc

```python
cdf['Dispacement'] = cars['Pojemność skokowa'].str.replace(' ', '').str.replace('cm3','').astype('int16')
```

7) Fuel type

```python
dummyF = pd.get_dummies(cars['Rodzaj paliwa'])
cdf['Diesel'] = dummyF['Diesel']
cdf['LPG'] = dummyF['Benzyna+LPG']
```

8) Body style: more than 3 doors

```python
cdf['Doors5'] = np.where(cars['Liczba drzwi']>3, 1, 0).astype('uint8')
```

9) First owner

```python
cdf['FirstOwner'] = cars['Pierwszy właściciel'].replace(['Tak', '0'], [1, 0]).astype('uint8')
```

10) Serviced at the Dealership

```python
cdf['ServDealership'] = cars['Serwisowany w ASO'].replace(['Tak', '0'], [1, 0]).astype('uint8')
```

11) Other features:
- ABS
- Electric windows
- Power steering
- Parking sensors
- Multifunctional steering
- Aluminium alloy wheels
- Cruise control
- On-board computer

```python
cdf['ABS'] = cars['ABS'].astype('uint8')
cdf['ElecWin'] = cars['Elektryczne szyby przednie'].astype('uint8')
cdf['PowSteer'] = cars['Wspomaganie kierownicy'].astype('uint8')
cdf['ParkingSens'] = cars['Czujniki parkowania tylne'].astype('uint8')
cdf['MultiSteer'] = cars['Wielofunkcyjna kierownica'].astype('uint8')
cdf['AAWheels'] = cars['Alufelgi'].astype('uint8')
cdf['CruContr'] = cars['ASR (kontrola trakcji)'].astype('uint8')
cdf['OnBoardComp'] = cars['Komputer pokładowy'].astype('uint8')
```

12) Country of origin

```python
dummyC = pd.get_dummies(cars['Kraj pochodzenia'])
cdf['Native'] = dummyC['Polska']
```

## 2.5 Futher data cleaning
----------------------

### 2.5.1 Price
I clean the price feature from the outlying ($3\sigma$) points for every value of age. The deviation may be caused by:

- typing mistake
- overvaluation of a car price
- selling a broken car

```python
def replace_incorrect_values(df):
    for i_age in range(28):
        data = df[df['Age'] == i_age]
        left = data['Price'].mean() - 3.* data['Price'].std()
        right = data['Price'].mean() + 3.* data['Price'].std()
        df = df.drop(data[(data['Price'] < left) | (data['Price'] > right)].index)
    return df
        
print('Before cleaning: ',len(cdf))
cdf = replace_incorrect_values(cdf)
print('After cleaning: ',len(cdf))
```

### 2.5.2 Mileage
There are still few cars with mistakes in data. Look at min, max values of Mileage, Dispcacement. Let's reject all cars with suspiciously low mileage.


```python
cdf.plot.scatter('Mileage', 'Age')
```

```python
cdf = cdf[(cdf['Age']-2.)*2000 < cdf['Mileage']]
```

There are two cars with mileage grater than million. It is probably mistake in typing. 

```python
cdf[cdf['Mileage'] > 5e5] = cdf[cdf['Mileage'] > 1e6]/10.
```

### 2.5.3 Dispacement
Opel Astra was produced with engines smaller than 2.2 L.

```python
cdf.plot.scatter('Dispacement', 'Age')
```

```python
cdf = cdf[cdf['Dispacement'].between(800, 2250)]
```
